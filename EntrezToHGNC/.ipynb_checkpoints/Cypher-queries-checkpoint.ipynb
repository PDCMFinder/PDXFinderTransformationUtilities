```python
%%bash
#get count of xenografts with specimens,molecularCharacterization, and markerAssociations
cypher-shell -u neo4j -p neo5j "match (mc:ModelCreation)-[:MODEL_SAMPLE_RELATION]-(s:Sample)-[:CHARACTERIZED_BY]-(mol:MolecularCharacterization)-[:ASSOCIATED_WITH]-(ma:MarkerAssociation) 
with mc,s,mol,ma
match (s)--(spe:Specimen)
return count(mc) as "xenografts_with_mc"";
```

```python
%%bash
#Count patient samples that have tumor types and tissues
cypher-shell -u neo4j -p neo5j "match (p:Patient)-[:COLLECTION_EVENT]-(ps:PatientSnapshot)-[:SAMPLED_FROM]-(s:Sample)-[:IMPLANTED_IN]-(mc:ModelCreation)
with p,ps,s,mc
match (s)-[:OF_TYPE]-(:TumorType)
with p,ps,s,mc
match (s)--(:Tissue)
return count(s) as "patient_samples_with_tt_and_tissue"";
```

```python
%%bash
#List all sample id's 
cypher-shell -u neo4j -p neo5j "match (p:Patient)-[:COLLECTION_EVENT]-(ps:PatientSnapshot)-[:SAMPLED_FROM]-(s:Sample)-[:IMPLANTED_IN]-(mc:ModelCreation)
with p,ps,s,mc
match (s)-[:OF_TYPE]-(:TumorType)
with p,ps,s,mc
match (s)--(:Tissue)
return s.sourceSampleId";
```
