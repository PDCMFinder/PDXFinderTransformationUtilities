```python
%%bash
rsync -avz /home/afollette/Finder_Data_Repositories/pdxfinder-data-release/data/UPDOG/PMLB/cna/PMLB_cna.tsv ./
```

```python
entrezToHugoFile = open("custom.csv", 'r')
cnaFile = open("PMLB_cna.tsv", 'r')
unmappedEntrez = open("unmappedEntrezList", "w+")
mapDf = pa.read_csv(entrezToHugoFile, sep='\t', keep_default_na=False).astype(str)
cnaDf = pa.read_csv(cnaFile, sep='\t',keep_default_na=False).astype(str)
```

```python
#test
row = cnaDf.iloc[1]
symbol = mapDf[mapDf.iloc[:, 1] == row['ncbi_gene_id']].iat[0,1]
print(type(symbol))
print(int(symbol))
```

```python
columnOrder = cnaDf.columns.tolist()
newCna = pa.DataFrame()
tmp = pa.DataFrame()
for index,row in cnaDf.iterrows():
    matchedRow = mapDf[mapDf.iloc[:, 1] == row['ncbi_gene_id']]
    if len(matchedRow) == 1 and not matchedRow.empty:
        row['symbol'] = matchedRow.iat[0,0]
        tmp = newCna.append(row, ignore_index=True)
        newCna = tmp
    else:
        unmappedEntrez.write("{}\n".format(row['ncbi_gene_id'])
        print("not mapped {}".format(row['ncbi_gene_id']))
```

```python
unmappedEntrez.flush()
unmappedEntrez.close()
fileDf = newCna[columnOrder].astype(str)
fileDf.to_csv("PMLB_cna_symbols.tsv", sep='\t', index=False)
```

```python
%%bash
rsync -v PMLB_cna_symbols.tsv working_cna.tsv
cut -f 11 working_cna.tsv | sort | uniq > postFilterEntrez 
cut -f 11 PMLB_cna.tsv > preFilterNotUniqEntrez 
sort preFilterNotUniqEntrez | uniq > preFilterEntrez

```

```python
%%bash
#There should not be any Entrez Id's in the post filter there were not
# in the prefilter
join -j 1 -v 2 preFilterEntrez postFilterEntrez
```

```python
%%bash
#Contains missing unmapped Entrez id's
join -j 1 -v 1 preFilterEntrez postFilterEntrez --check-order | sort -bg > unmappedEntrez
wc -l unmappedEntrez
```

```python
%%bash
#Grab count the occurances of the umapped entrez Id's
while IFS='' read -r LINE || [ -n "${LINE}" ]; do
    grep -c ${LINE} preFilterNotUniqEntrez >> entrezCount
done < unmappedEntrez
```

```python
%%bash
#sum total. Should account for lost data points. it does
awk '{count+=$1;} END{print count;}' entrezCount
```
