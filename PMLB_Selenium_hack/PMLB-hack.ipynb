```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
import pandas as pa
import time
```

```python
entrezFile = open("EntrezIdResume", "r")
#entrezFile = open("entrezIds", "r")
outFile = open("HSNCC_log2cna.json", "a+")
log = open("pmlb-hack-log", "a+")
entrezDf = pa.read_csv(entrezFile)
```

```python
driver = webdriver.Firefox()
driver.get("https://cbioportal.ca")
```

```python
#Please sign into cbioportal with the account with
# the correct permission
```

```python
for index, row in entrezDf.iterrows():
    geneId = row.iat[0]
    uri = "https://cbioportal.ca/cbioportal/api/molecular-profiles/HNSCC_Engraftment_log2CNA/molecular-data?sampleListId=HNSCC_Engraftment_all&entrezGeneId={}&projection=SUMMARY".format(geneId)
    driver.get(uri)
    logOut = ("{} \n entrezId: {} \n uri:{}".format(time.ctime(), geneId, uri ))
    print(logOut)
    log.write(logOut)
    time.sleep(4)
    rawButton = driver.find_element_by_id("rawdata-tab")
    rawButton.click()
    text = driver.find_elements_by_class_name("data")
    while len(text) is not 0:
        json = text.pop().text
        print("\n json length : {} \n".format(len(json)))
        log.write("\njson length : {}\n".format(len(json)))
        outFile.write(json)
        log.flush()
        outFile.flush()
```
