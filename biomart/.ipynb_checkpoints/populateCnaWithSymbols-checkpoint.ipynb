```python
import re 
import biomart as bt
import pandas as ps
```

```python
cnaDf = ps.read_csv("/home/afollette/Finder_Data_Repositories/pdxfinder-data-release-head/data_quarantine/UPDOG/UOC-BC/cna/UOC-BC_cna.tsv", sep='\t')
cnaDf["genomic_region"] = cnaDf["chromosome"].astype(str) + ":" + cnaDf["seq_start_position"].astype(str) + ":" + cnaDf["seq_end_position"].astype(str) + ":1"
genomicCoor = cnaDf["genomic_region"]
```

```python
coorSize = cnaDf['genomic_region'].size
```

```python
server = bt.BiomartServer( "http://grch37.ensembl.org/biomart" )
server.is_alive
gene = server.datasets["hsapiens_gene_ensembl"]
```

```python
for i in range(50,coorSize,50):  
    response = gene.search({ 
            'filters' : {
                'chromosomal_region' : genomicCoor[i-49:i].tolist()
            },
            'attributes': [
               'hgnc_symbol', 'chromosome_name', 'start_position', 'end_position' 
            ]
            })
    geneResponse = response.content.decode()
    geneResponse.replace('\n', '\t').split('\t')
```
