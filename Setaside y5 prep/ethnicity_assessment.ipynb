```python
from pymupdf.mupdf import pdf_is_bool

from utils import *
```

```python
df = pd.read_json("https://www.cancermodels.org/api/search_index?select=external_model_id,data_source,provider_name,model_type,patient_ethnicity,patient_ethnicity_assessment_method,raw_data_resources")
mm_df = pd.read_json("https://www.cancermodels.org/api/model_metadata?select=model_id,data_source,type,contact_name_list,contact_email_list")
```

```python
df['patient_ethnicity'] = df['patient_ethnicity'].str.lower().str.replace('not collected', 'not provided')
merged_df = df.merge(mm_df, left_on=['external_model_id', 'model_type', 'data_source'], right_on=['model_id', 'type', 'data_source'], how='left')
merged_df
```

```python
merged_df['patient_ethnicity'].unique()
```

```python
no_ethnicity = merged_df[merged_df['patient_ethnicity'] == 'not provided']
no_ethnicity
```

```python
grouped = no_ethnicity.groupby(['data_source', 'model_type']).count()['external_model_id']
pivot_table = grouped.unstack(fill_value=0)
```

```python
pivot_table.merge()
```

```python
result = (merged_df[['data_source', 'contact_name_list', 'contact_email_list']].fillna('Not provided').groupby("data_source", as_index=False)
    .agg({
        "contact_name_list": lambda x: ", ".join(sorted(set(x))),
        "contact_email_list": lambda x: ", ".join(sorted(set(x))),
    })
)

result
```

```python
pivot_table.reset_index().merge(result, how='left', on='data_source')
```

```python
ethnicity = merged_df[merged_df['patient_ethnicity'] != 'not provided']
ethnicity
```

```python
(ethnicity.fillna('Not provided').groupby("data_source", as_index=False)
    .agg({
        "patient_ethnicity_assessment_method": lambda x: ", ".join(sorted(set(x))),
    })
)
```

```python
(merged_df.fillna('').groupby("data_source", as_index=False)
    .agg({
        "raw_data_resources": lambda x: sorted(set(item for sublist in x for item in sublist))
    })
)
```
