```python
import pandas as ps
```

```python
masterList = open('completedEnaSamplesMetadata.csv','r')
fastqInfo = open('fastqMetadataPdmrWes_filled.tsv', 'r')
```

```python
samplesDf = ps.read_csv(masterList, sep='\t')
fastqDf = ps.read_csv(fastqInfo, sep='\t')
```

```python
unmatched = samplesDf[(~samplesDf.sample_alias.isin(fastqDf.sample_alias))]
```

```python
unmatched.shape
```

```python
nonOrig = unmatched[~unmatched.sample_alias.str.contains('-[RT].+-[RT]', regex=True)]
```

```python
nonOrig.to_csv("notMatching.tsv", sep='\t')
```
